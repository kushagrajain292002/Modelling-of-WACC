```


# List of tickers (Yahoo Finance format).
# For Indian stocks use the exchange suffix (.NS for NSE, .BO for BSE)
tickers = [
    "DMART.NS",             # Avenue Supermarts (DMart)
    "TRENT.NS",             # Trent
    "MANYAVAR.NS",          # Vedant Fashions (NSE)
    "ABFRL.NS",             # Aditya Birla Fashion and Retail
    "METROBRAND.NS"          # Metro Brands
]

output = []

for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)

        info = stock.info  # dictionary of fields

        market_cap = info.get("marketCap")
        total_debt = info.get("totalDebt")
        debt_to_equity = info.get("debtToEquity")

        output.append({
            "Ticker": ticker,
            "Market Cap": market_cap,
            "Total Debt": total_debt,
            "Debt to Equity": debt_to_equity
        })
    except Exception as e:
        output.append({
            "Ticker": ticker,
            "Market Cap": None,
            "Total Debt": None,
            "Debt to Equity": None,
            "Error": str(e)
        })

# Convert to DataFrame
df = pd.DataFrame(output)

# Write to Excel
file_name = "company_financials.xlsx"
df.to_excel(file_name, index=False)

print(f"Saved data to {file_name}")
```

```

# Date range
start_date = "2024-02-05"
end_date = "2026-02-06"

# Stock tickers
tickers = {
    "TRENT": "TRENT.NS",
    "MANYAVAR": "MANYAVAR.NS",
    "ABFRL": "ABFRL.NS",
    "ADN": "ADANIENT.NS",
    "METROBRAND": "METROBRAND.NS"
}

# Empty DataFrame to store closing prices
combined_data = pd.DataFrame()

# Download and combine
for name, ticker in tickers.items():
    data = yf.download(
        ticker,
        start=start_date,
        end=end_date,
        interval="1wk"
    )

    combined_data[name] = data["Close"]

# Remove rows with all NaN values (if any)
combined_data.dropna(how="all", inplace=True)

# Export to Excel
file_name = "Retail_Stocks_Weekly_Closing_Prices.xlsx"
combined_data.to_excel(file_name)

print("Excel file created successfully:", file_name)
```

```

# Download data
nifty = yf.download("^NSEI", start="2000-01-01", end="2026-01-01")

# Remove multi-level column index
nifty.columns = nifty.columns.get_level_values(0)

# Keep Close price
nifty = nifty[['Close']]

# Resample to year-end
yearly_prices = nifty.resample('YE').last()

# Calculate yearly return
yearly_prices['Yearly Return (%)'] = yearly_prices['Close'].pct_change() * 100

# Drop first NaN
yearly_prices = yearly_prices.dropna()

# Add Year column
yearly_prices['Year'] = yearly_prices.index.year

# Rearrange columns
final_data = yearly_prices[['Year', 'Close', 'Yearly Return (%)']]

# Save to Excel
final_data.to_excel("Nifty_Yearly_Returns_2000_2025.xlsx", index=False)

print("Excel file created successfully!")
```
